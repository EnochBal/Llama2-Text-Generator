```python
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q8_0.bin"
```

```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

```python
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2,
    n_batch=2048,
    n_gpu_layers=32
    )
```

```python
lcpp_llm.params.n_gpu_layers
```

```python
prompt = "Write a story with a minimum of 1000 words length"
prompt_template=f'''SYSTEM: You are a brilliant storyteller.

USER: {prompt}

ASSISTANT:
'''
```

```python
response=lcpp_llm(prompt=prompt_template, max_tokens=1000000, temperature=0.7, top_p=0.95,
                  repeat_penalty=1.2, top_k=150,
                  echo=True)
```
